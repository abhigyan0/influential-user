## Graphs

### Tweet vs Retweets

```python
sports_data['row_id'] = range(1, len(sports_data) + 1)
x_axis = sports_data["row_id"]
y_axis = sports_data["retweets"]
plt.plot(x_axis,y_axis)
plt.xlabel('Tweet Id')
plt.ylabel('Retweets')
plt.title('Tweet vs Retweets')
plt.show()
```

### Tweet Id vs Favourite

```python
x_axis = sports_data["row_id"]
y_axis = sports_data["favorites"]

plt.plot(x_axis,y_axis)
plt.xlabel('Tweet Id')
plt.ylabel('favorites')
plt.title('Tweet vs favorites')
plt.show()
```

### Tweet Id vs Polarity

```python
x_axis = sports_data["retweets"]
y_axis = sports_data["favorites"]

plt.plot(x_axis,y_axis)
plt.xlabel('retweets')
plt.ylabel('favorites')
plt.title('retweets vs favorites')
plt.show()
```

### Negative vs positive sentiments:

```python
sentiment_data = sports_data["sentiment"]
positive_count = 0
negative_count = 0
for i in sentiment_data:
    if i == "pos":
        positive_count = positive_count + 1
    if i == "neg":
        negative_count = negative_count + 1
    else:
        pass
pos_prcnt = int((positive_count/(positive_count+negative_count)) * 100)
neg_prcnt = int((negative_count/(positive_count+negative_count)) * 100) + 1 
labels = 'Positive', 'Negative'
explode = (0, 0.1) 
sizes = [pos_prcnt,neg_prcnt]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Top Influencers

```python
top_tweets = sports_data.sort_values("retweets", axis = 0, ascending = False)
#print(top_tweets)
name = []
retweet = []
for index, row in top_tweets.iterrows():
    if len(retweet) < 10:
        if row[9] != "NaN":
            d = str((row[9]))
            y = d.split("/")
            if (len(y)) == 6:
                name.append(y[3])
                retweet.append(row[3])
            else:

                pass
plt.bar(name, retweet)
plt.grid()
plt.title('Top Influencers')
plt.ylabel('Retweet')
plt.xticks(rotation='vertical')
plt.show()
```

## Comparing Different models

```python
title = ['TF_IDF', 'Count Vectorization', 'Logistic Regression', 'Naive Bayes', 'SVC']
values = list((model_data.values()))
dec_values = []
for i in values:
    c=round(i,2)
    dec_values.append(c)
plot_values = dec_values
plt.bar(title, dec_values)
plt.grid()
plt.title('Accuracy')
plt.ylabel('Accuracy (%)')
plt.xticks(rotation='vertical')
for index,data in enumerate(dec_values):
    plt.text(x=index , y =data , s=f"{data}")
#chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.show()
```

#### Above graph represent text based accuracy comparision over different classification techniques.It was found that accuracy was highest for the Naive Bayes.
